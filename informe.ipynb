## Primera parte

### a.
El Espacio Col(X) es el conjunto de todos los vectores de ℝⁿ que pueden escribirse como combinación lineal de las columnas de X.

¿Col(X) es un subespacio vectorial de ℝⁿ?

Col(X) = {b en ℝⁿ | b = X·β con β variando en ℝᵖ}

- ¿0 pertenece a Col(X)?

Consideramos el vector β = 0 en ℝᵖ:  
⇒ b = X·β = X·0 = 0 ∈ ℝⁿ  
⇒ 0 ∈ Col(X)

- Dado b₁ y b₂ ∈ Col(X), ¿b₁ + b₂ ∈ Col(X)?

Por definición existen β₁, β₂ ∈ ℝᵖ tales que: b₁ = X·β₁ y b₂ = X·β₂  
Verificamos la suma: b₁ + b₂ = X·β₁ + X·β₂ = X·(β₁ + β₂)  
Como β₁ + β₂ ∈ ℝᵖ, entonces b₁ + b₂ ∈ Col(X)

- Dado b ∈ Col(X) y c ∈ ℝ, ¿c·b ∈ Col(X)?

⇒ b = X·β ⇒ c·b = c·(X·β) = X·(c·β)  
Entonces, c·b ∈ Col(X)

Verificamos que Col(X) contiene al vector 0, b₁ + b₂ y c·b. Por lo tanto, se cumplen todas las condiciones para ser un subespacio de ℝⁿ.

### b.
Dados los vectores u, v ∈ ℝⁿ, queremos ver que u·v = vᵗ·u

u ∈ ℝⁿˣ¹, v ∈ ℝⁿˣ¹, vᵗ ∈ ℝ¹ˣⁿ

- u·v = Σ (uᵢ·vᵢ)
- vᵗ·u = (v₁, ..., vₙ) · u = Σ (vᵢ·uᵢ)

Cada término vᵢ·uᵢ es un producto de escalares reales, y en ℝ la multiplicación es conmutativa. Entonces: vᵢ·uᵢ = uᵢ·vᵢ lo que muestra que Σ (uᵢ·vᵢ) = Σ (vᵢ·uᵢ). Por lo que vᵗ·u = u·v

### c.
Definición de solución óptima: ‖y - X·β*‖ = min ∥y − Xβ∥ , β en ℝᵖ

Aplicamos el teorema de la proyección ortogonal:

- Tomamos S = Col(X), vector y ∈ ℝⁿ, vector en S b = X·β*
- El teorema afirma que minimizar la distancia equivale a (y - b)·s = 0, para todo s en S

Sustituimos b = X·β*: (y - X·β*) · s = 0 , para todo s ∈ Col(X)

Dado que cada s ∈ Col(X) puede escribirse como s = X·β con β ∈ ℝᵖ, la condición queda: (y - X·β*) · X·β = 0

Así logramos traducir la minimización de la norma ‖y - X·β*‖ en la condición de ortogonalidad.

### d.
Queremos llegar a: Xᵗ(y - X·β*) = 0

- Partimos de la ecuación obtenida en el inciso (c):
  (y - X·β*) · (X·β) = (X·β)ᵗ · (y - X·β*)
- Reescribimos: (X·β)ᵗ · (y - X·β*) = βᵗXᵗ(y - X·β*)

La condición (c) se transforma en (X·β)ᵗ · (y - X·β*) = 0, que a su vez se transforma en Xᵗ(y - X·β*)·β = 0, que es la ecuación solicitada.

### e.
Queremos llegar a la fórmula: Xᵗ·X·β* = Xᵗ·y

- Partimos de: Xᵗ(y - X·β*)·β = 0
- Definimos el vector u: u = Xᵗ(y - X·β*) ∈ ℝᵖ ⇒ La condición anterior queda como u·β = 0
- En ℝᵖ, el único vector u que satisface u·v = 0 para todo v es el vector nulo u = 0 (enunciado)
- Entonces, de u = 0 → Xᵗ(y - X·β*) = 0 → Xᵗy - XᵗX·β* = 0
- Reordenamos la ecuación anterior: XᵗX·β* = Xᵗy, llegamos a la ecuación solicitada

### f.
Queremos llegar a la fórmula de la solución óptima al problema de regresión:

- Partimos de las ecuaciones normales: XᵗX·β* = Xᵗy
- Suponemos que las columnas de X son linealmente independientes. En ese caso, la matriz XᵗX (tamaño p×p) es invertible.
- Multiplicamos ambos lados de la ecuación por (XᵗX)^-1:
  (XᵗX)^-1·XᵗX·β* = (XᵗX)^-1·Xᵗy
- (XᵗX)^-1·(XᵗX) = Iₚ (la identidad en ℝᵖ). Obtenemos: Iₚ·β* = (XᵗX)^−1·Xᵗy

⇒ β* = (XᵗX)^−1·Xᵗy

β* es el vector de coeficientes que minimiza la suma de los cuadrados de los residuos.
